 # Practical Data Visualization
## Creating standard data graphics


```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
import requests

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Creating a line chart from a list object

#### Plotting a line chart in matplotlib

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

#### Plotting a line chart from a Pandas object

```python

cars = pd.read_csv("https://raw.githubusercontent.com/asel-datascience/ODS2020/master/Week_5_Data_Visualization_Python/mtcars.csv")

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars['mpg']
```

```python
mpg.plot()
```

```python
df = cars[['cyl','wt','mpg']]
df.plot()
```

### Creating bar charts
#### Creating a bar chart from a list

```python
plt.bar(x, y)
```

#### Creating bar charts from Pandas objects

```python
mpg.plot(kind="bar")
```

```python
mpg.plot(kind="barh")
```

### Creating a pie chart

```python
x = [1,2,3,4,0.5]
plt.pie(x)
plt.show()
```

### Saving a plot

```python
plt.pie(x)
plt.savefig('pie_chart.png')
plt.show()
```

# Defining elemnts of a plot 

```python
%matplotlib inline
rcParams['figure.figsize']= 5,4
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```
